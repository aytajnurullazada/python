```python
import numpy as np
from numpy.linalg import matrix_power
```

```python
fl = np.array([
    [.1,.8,.6],
    [.5,.7,.5],
    [.2,.2,.8]
])
layer = np.random.rand(6,6)
layer = layer.round(1)

def manual_conv2D(layer, stride, fl, padding):
    f_size = len(fl)
    rows, cols = np.shape(layer) #layerin olculerini aliriq
    
    #padding tetbiq edirik

    padded_layer = np.zeros((rows + (padding*2), cols + (padding*2)))
    padded_layer[padding: rows+padding, padding: cols+padding] = layer

    # print(layer)
    # print(padded_array)

    #padding tetbiq olunmush layerin yeni olchulerini aliriq
    rows, cols = np.shape(padded_layer)

    #conv
    r = int((rows - f_size)/stride + 1) # yeni layerin row sayisi
    cl = int((cols - f_size)/stride + 1) # yeni layerin column sayisi

    conv2d_layer = np.zeros((r, cl))
    
    for i in range(0, r, stride):
        for j in range(0, cl, stride):
            res = round(np.sum(np.multiply(padded_layer[i: i+f_size, j: j+f_size], fl)), 2)
            conv2d_layer[i][j] = res
    
    return conv2d_layer


result = manual_conv2D(layer, 1, fl, 1)
# print(np.shape(result))
# print(result)
```

```python
# Pooling method
def manual_pooling(layer, stride, kernel):
    rows, cols = np.shape(layer) #layerin olculeri

    r = int((rows - kernel)/stride + 1) 
    cl = int((cols - kernel)/stride + 1) 
    
    pooled_layer = np.zeros((r, cl))
    
    for i in range(r):
        for j in range(cl):
            res = np.amax(layer[i:i+kernel, j:j+kernel]) #max pooling
            # res = np.amin(layer[i:i+kernel, j:j+kernel]) #min pooling
            # res = round(np.average(layer[i:i+kernel, j:j+kernel]), 2) #average pooling
            pooled_layer[i][j] = res

    return pooled_layer

# print(layer)
# print(manual_pooling(layer, 1, 3)) 


# Flatten method

# layer = np.random.rand(10, 10)
# layer = layer.round(1)


def manual_flatten(matrix):
    res = np.array([])
    for i in range(len(matrix)):
        res = np.insert(res, len(res), matrix[i])
    return res

# print(manual_flatten(layer))
```
